```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Best Cities for Startups.csv.xls")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df2 = df.head(50)
df2
```

```python
df.shape
```

```python
plt.scatter(df2["quality score"], df2["country"])
plt.xlabel('quality score')
plt.ylabel('country')
plt.grid()
plt.savefig("country quality score.jpg", bbox_inches='tight',pad_inches= 1)
plt.show()
```

```python
plt.scatter( df2["business score"], df2["country"])
plt.xlabel('business score')
plt.ylabel('country')
plt.grid()
plt.savefig("country business score.jpg", bbox_inches='tight',pad_inches= 1)
plt.show()
```
